Copo de nieve por la curva de Koch
Fractales: figuras geométricas que se pueden fraccionar en partes que son o se asemejan una versión reducisa de la figura completa
Curva de Koch   El creador en 1904 de este monstruo fue Niels Fabian Helge von Koch , matemático sueco.   Partamos de un triángulo equilátero de lado unidad. Dividimos en tres partes iguales de longitud 1/3 cada lado. Sustituimos el segmento central por dos segmentos de tamaño idéntico formando un diente como muestra la animación en la iteración n=1.  Tenemos una curva poligonal P1 de longitud 3·4··1/3=4. Repetimos la operación (n=2) con cada uno de los cuatro nuevos segmentos de cada uno de los "lados". Obtendremos así la curva P2 de longitud 3·42·1/32=16/3. La iteración indefinida nos proporciona la isla de Koch o copo de nieve de Koch.
